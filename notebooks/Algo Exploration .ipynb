## Exploring Betting Strategies using historical odds

Datasource 1: https://www.indatabet.com/ft-free-2in1.html

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data1 = pd.read_csv('football_historical.csv', low_memory = False)
```

```python
data1.head()
```

```python
data1.columns
```

```python
#Take Relevant Data
hist_data = pd.DataFrame()


hist_data['dates'] = data1['Date'][2:]
hist_data['ID'] = data1['ID (FIFA)'][2:] #identifies both country and league
hist_data['season'] = data1['Season'][2:]
hist_data['home_team'] = data1['Teams'][2:]
hist_data['away_team'] = data1['Unnamed: 9'][2:]
hist_data['result'] = data1['Unnamed: 12'][2:] #H - home wins, D - Draw, A - away wins
hist_data['pinnacle_home_odds'] = data1['FT Moneyline'][2:]
hist_data['pinnacle_draw_odds'] = data1['Unnamed: 24'][2:]
hist_data['pinnacle_away_odds'] = data1['Unnamed: 25'][2:]
hist_data['bet365_home_odds'] = data1['FT Moneyline.1'][2:]
hist_data['bet365_draw_odds'] = data1['Unnamed: 28'][2:]
hist_data['bet365_away_odds'] = data1['Unnamed: 29'][2:]

hist_data.head()
```
